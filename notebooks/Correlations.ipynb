```python
import numpy as np
from tensorflow.keras.models import Model, load_model
import tensorflow.keras.backend as K
from scipy.stats import pearsonr, spearmanr
import sys, os
sys.path.append('../python_scripts/')
import analysis, datasets, baseline
sys.path.append('../imported_code/svcca/')
import cca_core, pwcca
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

'''
Plot Helpers
'''
def color_plot_helper(title: str, corr_distance, axis):
    axis.set_title(title, fontsize=16)
    axis.set_xlabel('layer')
    axis.set_xticks(ticks=[5, 15, 25, 35, 45, 55, 65, 75, 85])
    axis.set_xticklabels([1, 2, 3, 4, 5, 6, 7, 8, 9])
    axis.set_yticks(ticks=[5, 15, 25, 35, 45, 55, 65, 75, 85])
    axis.set_yticklabels([1, 2, 3, 4, 5, 6, 7, 8, 9])
    axis.set_ylabel('layer')
    
    im = axis.imshow(corr_distance)
    plt.colorbar(im, ax=axis)
    
def mds_plot_helper(title: str, corr_distance, axis):
    embedding = MDS(n_components=2, dissimilarity='precomputed')
    X_transformed = embedding.fit_transform(corr_distance)
    
    axis.set_title(title, fontsize=16)
    colors = ['#f5f59d', '#f5b642', '#f58742', '#f55442', '#c940c5',
              '#b407ed', '#7d09d6', '#50017a', '#000000']
    for i in range(10):
        axis.set(adjustable='box', aspect='equal')
        axis.plot(X_transformed[i::10, 0], X_transformed[i::10, 1], linestyle='-', color='black', zorder=1)
        axis.scatter(X_transformed[i::10, 0], X_transformed[i::10, 1], marker='o', color=colors, zorder=2)
```

```python
'''
Load Data
'''
_, testData = datasets.make_train_data(None)
x_predict, _ = datasets.make_predict_data(testData)
```

```python
'''
Calculate RSA
'''
rsa_weights = analysis.correlate('RSA', '../outputs/models/primary/', x_predict)
rsa_shuffle = analysis.correlate('RSA', '../outputs/models/shuffle_seed/', x_predict)
# rsa_both = ...
```

```python
'''
Calculate SVCCA
'''
svcca_weights = analysis.correlate('SVCCA', '../outputs/models/primary/', x_predict)
svcca_shuffle = analysis.correlate('SVCCA', '../outputs/models/shuffle_seed/', x_predict)
```

```python
'''
Calculate PWCCA
'''
pwcca_weights = analysis.correlate('PWCCA', '../outputs/models/primary/', x_predict)
pwcca_shuffle = analysis.correlate('PWCCA', '../outputs/models/shuffle_seed/', x_predict)
```

```python
'''
Save weights
'''
np.save('pwcca_weights.np', pwcca_weights)
np.save('pwcca_shuffle.np', pwcca_shuffle)
np.save('svcca_weights.np', svcca_weights)
np.save('svcca_shuffle.np', svcca_shuffle)
np.save('rsa_weights.np', rsa_weights)
np.save('rsa_shuffle.np', rsa_shuffle)
```

```python
'''
Visualize RSA
'''
cf, (cax1, cax2, cax3) = plt.subplots(1, 3)
cf.suptitle('Color Graphs', fontsize=24)
cf.set_size_inches((20, 5.5))
color_plot_helper('Weights', 1-rsa_weights, cax1)
color_plot_helper('Shuffle', 1-rsa_shuffle, cax2)

mf, (max1, max2, max3) = plt.subplots(1, 3)
mf.suptitle('MDS', fontsize=24)
mf.set_size_inches((20, 5.5))
mds_plot_helper('Weights', 1-rsa_weights, max1)
mds_plot_helper('Shuffle', 1-rsa_shuffle, max2)
```
