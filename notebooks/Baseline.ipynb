```python
baseline.visualize_transform('shift', 16, imgset[105])
```

```python
dim = 32
num_imgs = 1000
versions = 1
model = models[0]
rep_orig = model.predict(imgset)
rep_orig = baseline.get_rdm(rep_orig)
correlate_func = baseline.do_rsa
correlations = []
up_imgset = np.zeros((num_imgs, dim, dim, 3), dtype=np.uint8)
down_imgset = np.zeros((num_imgs, dim, dim, 3), dtype=np.uint8)
left_imgset = np.zeros((num_imgs, dim, dim, 3), dtype=np.uint8)
right_imgset = np.zeros((num_imgs, dim, dim, 3), dtype=np.uint8)
# create the gray values we use to fill in space
empty = np.zeros((dim, dim, 3)).astype(np.uint8)
empty.fill(128)
for v in range(versions):
    # Generate transformed imageset
    print(' - Working on version', v, 'of', versions)
    for i in range(num_imgs):
        img = imgset[i, :, :, :]
        up_imgset[v]    = np.concatenate([img[v:dim, :, :], empty[0:v, :, :]])
        down_imgset[v]  = np.concatenate([empty[0:v, :, :], img[0:dim - v, :, :]])
        left_imgset[v]  = np.concatenate([img[:, v:dim, :], empty[:, 0:v, :]], axis=1)
        right_imgset[v] = np.concatenate([empty[:, 0:v, :], img[:, 0:dim - v, :]], axis=1)

    # Get average of all 4 directions
    print(' - Now correlating...')
    corr_sum = 0.
    rep = model.predict(up_imgset, verbose=1)
    corr_sum += correlate_func(rep_orig, rep)
    rep = model.predict(down_imgset, verbose=1)
    corr_sum += correlate_func(rep_orig, rep)
    rep = model.predict(left_imgset, verbose=1)
    corr_sum += correlate_func(rep_orig, rep)
    rep = model.predict(right_imgset, verbose=1)
    corr_sum += correlate_func(rep_orig, rep)
    correlations.append(corr_sum / 4)
```
